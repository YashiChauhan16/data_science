```python
#assignment1
#q-1
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(1, 100, size=(6, 4)), columns=['A', 'B', 'C', 'D'])
print("Original DataFrame:")
print(df)

df.set_index('A', inplace=True)
print("New index:")
df
```

```python
#q-2
df = pd.DataFrame(np.random.randint(1, 100, size=(3, 3)), columns=['A', 'B', 'C'], index=['X', 'Y', 'Z'])
print("DataFrame:")
print(df)

element = df.at['Y', 'B']
print("Row 'Y' and Column 'B':", element)
```

```python
#assignment2
#q-1
df = pd.DataFrame(np.random.randint(1, 100, size=(5, 3)), columns=['A', 'B', 'C'])
print("DataFrame:")
print(df)

df['D'] = df['A'] * df['B']
print("New DataFrame:")
print(df)
```

```python
#q-2
df = pd.DataFrame(np.random.randint(1, 100, size=(4, 3)), columns=['A', 'B', 'C'])
print("Original DataFrame:")
print(df)

sum1= df.sum(axis=1)
sum2= df.sum(axis=0)

print("Row sum:")
print(sum1)
print("Column sum:")
print(sum2)
```

```python
#assignment3
#q-1
df = pd.DataFrame(np.random.randint(1, 100, size=(5, 3)), columns=['A', 'B', 'C'])
print("Original DataFrame:")
print(df)
df.iloc[0, 1] = np.nan
df.iloc[2, 2] = np.nan
df.iloc[4, 0] = np.nan
print("NaN:")
print(df)
df.fillna(df.mean(), inplace=True)
print("Filled with mean:")
print(df)
```

```python
#q-2
df = pd.DataFrame(np.random.randint(1, 100, size=(6, 4)), columns=['A', 'B', 'C', 'D'])
print("Og df:")
print(df)
df.iloc[1, 2] = np.nan
df.iloc[3, 0] = np.nan
df.iloc[5, 1] = np.nan
print("NaN:")
print(df)
df.dropna(inplace=True)
print("New df:")
print(df)
```
