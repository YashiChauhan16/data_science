File operations and Exception handling.

```python
#Q-1
def file(file_name):
    with open(file_name,"rt")as f:
        for line in f:
            print(line.strip())

file("sample.txt")
```

```python
#Q-2
def write_list(lines,finlename):
    with open("filename","w")as a:
        for line in lines:
            a.write(line + "\n" )

write_list(["My","Name","is","Yashi."],"output.txt")
```

```python
#Q-3
def copy_text(s, d):
    with open(s, 'r') as source:
        with open(d, 'w') as destination:
            destination.write(source.read())

copy_text("source.txt","destination.txt")
```

```python
#Q-4
def appending(filename,a):
    with open(filename,"a")as file:
        file.write(a)
       

appending("log.txt","\nThis a new line.")
```

```python
#Q-5
def count_words(filename):
     with open(filename,"r")as document:
        text=document.read()
        words=text.split()
        return len(words)


count_words("document.txt")
       
```

```python
#Q-6
def replacing(filename,old , new):
    with open(filename, 'r') as file:
        text = file.read()
    new_text = text.replace(old, new)
    with open(filename, 'w') as file:
        file.write(new_text)


replacing("data.txt","Text","Binary")
```

```python
#Q-7
def reverse(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    for line in reversed(lines):
        print(line.strip())

reverse("reverse.txt")
```

```python
#Q-8
def count_lwc(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        words = sum(len(line.split()) for line in lines)
        characters = sum(len(line) for line in lines)
    return len(lines), words, characters

count_lwc("stats.txt")
```

```python
#Q-9
def merging(file_list, output_file):
    with open(output_file, 'w') as outfile:
        for fname in file_list:
            with open(fname, 'r') as infile:
                outfile.write(infile.read() + '\n')


merging(["data.txt","log.txt"],"merged.txt")
```

```python
#Q-10
def seprate(filename, lines_per_file):
    with open(filename, 'r') as file:
        lines = file.readlines()
    for i in range(0, len(lines), lines_per_file):
        with open(f'{filename}_part{i//lines_per_file + 1}.txt', 'w') as part_file:
            part_file.writelines(lines[i:i + lines_per_file])


seprate('large.txt', 100)
```

```python
#Q-11
import datetime

def log_file(message, filename='activity.log'):
    timestamp = datetime.datetime.now().isoformat()
    with open(filename, 'a') as file:
        file.write(f'[{timestamp}] {message}\n')


log_file('This is a log message.')
```

```python
#Q-12
def binary_file(source, destination):
    with open(source, 'rb') as src:
        with open(destination, 'wb') as dest:
            dest.write(src.read())

binary_file('image.bin', 'copy_image.bin')
```

```python
#Q-13
import csv
def csv_file(filename):
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        return list(reader)


print(csv_file('data.csv'))
```

```python
#Q-14
import json

def read_json(filename):
    with open(filename, 'r') as file:
        data = json.load(file)
        return data


print(read_json('data.json'))
```

```python
#Q-15
def permission(filename):
    try:
        with open(filename, 'r') as file:
            print(file.read())
    except PermissionError as e:
        print(f"Permission error: {e}")


permission('protected.txt')
```
